```python
import pandas as pd

df = pd.read_csv('bank.csv')
df['Date'] = pd.to_datetime(df['Date'])
print(df.head())
```

```python
df = df[['Date', 'Running Bal.']]
print(df.head())
```

```python
monthly = df.groupby(pd.Grouper(key='Date', freq='ME')).agg({'Running Bal.': 'mean'})
monthly['Previous Mo'] = monthly['Running Bal.'].shift(1)
monthly['Owners Draw'] = round(((26*3500)/12)+846, 2)
monthly['Implied Profit'] = monthly['Running Bal.'] - monthly['Previous Mo'] + monthly['Owners Draw']
monthly['Prior Year'] = monthly['Implied Profit'].shift(12)
monthly['YoY'] = monthly['Implied Profit'] / monthly['Prior Year']

print(monthly)
```
